```python
import pandas as pd 
pass_tit=pd.read_csv("titanic-passengers.csv",sep=";")
pass_tit
```

```python
nom=pass_tit["Name"]
nom
```

```python

Title_Dictionary = {

                    "Capt.":       "Officer",

                    "Col.":        "Officer",

                    "Major.":      "Officer",

                      "Dr.":      "Officer",

                    "Rev.":        "Officer",

                    "Jonkheer.":   "Royalty",

                    "Don.":        "Royalty",

                    "Sir." :       "Royalty",

                   "Lady." :      "Royalty",

                  "the Countess.": "Royalty",

                    "Dona":       "Royalty",

                    "Mme.":        "Miss",

                    "Mlle.":       "Miss",

                    "Miss." :      "Miss",

                    "Ms.":         "Mrs",

                    "Mr." :        "Mr",

                    "Mrs." :       "Mrs",

                    "Master." :    "Master"

                    }

```

# changing the titles in name using the Title dictionary

```python
print(nom)
nom=nom.values.tolist()

L=["Master",'Mr' ,"Mrs", "Miss", "Royalty", "Officer"]

for index, data in enumerate(nom):
         for key, value in Title_Dictionary.items():
                  if key in data:
                            nom[index]=data.replace(key, Title_Dictionary[key])

                        
```

# creating the list of tiltles 

```python
def findinl(str,L):
    str=str.split(' ')
    for i in range(0,len(str)):
        if str[i] in L:
            return str[i]

Titles=[]

for n in range(0,len(nom)):
    x=findinl(nom[n],L)
    Titles.append(x)
Titles
```


# creating the new dataframe 

```python
names_df=pd.DataFrame(nom)
Title=pd.DataFrame(Titles)
Title
names_df.loc[:, "title"] = pd.Series(Title[0], index=names_df.index)
names_df.loc[:, "Age"] = pd.Series(pass_tit["Age"], index=names_df.index)
names_df.loc[:, "Sex"] = pd.Series(pass_tit["Sex"], index=names_df.index)
names_df.loc[:, "Fare"] = pd.Series(pass_tit["Fare"], index=names_df.index)
names_df.loc[:, "Sibsp"] = pd.Series(pass_tit["SibSp"], index=names_df.index)
names_df.loc[:, "Parch"] = pd.Series(pass_tit["Parch"], index=names_df.index)
print(names_df)
```

```python
namcol=names_df.columns
namcol
names_df.columns=["name", 'title', 'Age', 'Sex', 'Fare', 'Sibsp', 'Parch']
names_df["familysize"]=  names_df['Sibsp'] + names_df['Parch']+1
names_df
```
